```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter1d

# Load results
results_csv = '/content/yolov5/runs/train/exp/results.csv'
df = pd.read_csv(results_csv)
df.columns = df.columns.str.strip()

# Extract mAP@0.5 (proxy for accuracy)
train_map = df['metrics/mAP_0.5']
val_map = df['metrics/mAP_0.5']  # Same column, but validation is post-train eval

# Apply Gaussian filter
sigma = 1  # Smoothing factor
train_map_smooth = gaussian_filter1d(train_map, sigma=sigma)
val_map_smooth = gaussian_filter1d(val_map, sigma=sigma)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(train_map, label='Training mAP@0.5 (Raw)', alpha=0.3)
plt.plot(train_map_smooth, label='Training mAP@0.5 (Smoothed)', linewidth=2)
plt.plot(val_map, label='Validation mAP@0.5 (Raw)', alpha=0.3)
plt.plot(val_map_smooth, label='Validation mAP@0.5 (Smoothed)', linewidth=2)
plt.xlabel('Epoch')
plt.ylabel('mAP@0.5')
plt.title('Training vs Validation Accuracy (mAP@0.5)')
plt.legend()
plt.grid()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter1d

# Load results
results_csv = '/content/yolov5/runs/train/exp/results.csv'
df = pd.read_csv(results_csv)
df.columns = df.columns.str.strip()

# Calculate total loss (train and val)
train_loss = df['train/box_loss'] + df['train/obj_loss'] + df['train/cls_loss']
val_loss = df['val/box_loss'] + df['val/obj_loss'] + df['val/cls_loss']

# Apply Gaussian filter
sigma = 1
train_loss_smooth = gaussian_filter1d(train_loss, sigma=sigma)
val_loss_smooth = gaussian_filter1d(val_loss, sigma=sigma)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(train_loss, label='Training Loss (Raw)', alpha=0.3)
plt.plot(train_loss_smooth, label='Training Loss (Smoothed)', linewidth=2)
plt.plot(val_loss, label='Validation Loss (Raw)', alpha=0.3)
plt.plot(val_loss_smooth, label='Validation Loss (Smoothed)', linewidth=2)
plt.xlabel('Epoch')
plt.ylabel('Total Loss')
plt.title('Training vs Validation Loss')
plt.legend()
plt.grid()
plt.show()
```

```python
import os
import numpy as np
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Run validation and save predictions
!python val.py --weights /content/yolov5/runs/train/exp/weights/best.pt --data /content/drive/MyDrive/yolov5_dataset/data.yaml --img 640 --task test --save-txt --save-conf

# Paths
label_dir = '/content/drive/MyDrive/yolov5_dataset/test/labels'  # Ground truth
pred_dir = '/content/yolov5/runs/val/exp2/labels'  # Predictions

# Initialize lists
y_true = []
y_pred = []

# Class mapping
class_names = ['Boulder', 'Crater']  # Only these two
valid_classes = [0, 1]  # Class IDs to keep

# Load ground truth
for label_file in os.listdir(label_dir):
    with open(os.path.join(label_dir, label_file), 'r') as f:
        for line in f:
            class_id = int(line.split()[0])
            if class_id in valid_classes:  # Filter out Plain surface (2)
                y_true.append(class_id)

# Load predictions
for pred_file in os.listdir(pred_dir):
    with open(os.path.join(pred_dir, pred_file), 'r') as f:
        for line in f:
            class_id = int(line.split()[0])
            conf = float(line.split()[-1])
            if class_id in valid_classes and conf > 0.25:  # Confidence threshold
                y_pred.append(class_id)

# Ensure lengths match (truncate to shorter)
min_len = min(len(y_true), len(y_pred))
y_true = y_true[:min_len]
y_pred = y_pred[:min_len]

# Compute confusion matrix
cm = confusion_matrix(y_true, y_pred, labels=valid_classes)

# Calculate overall accuracy
true_positives = cm[0, 0] + cm[1, 1]  # TP for Boulder + TP for Crater
total_predictions = cm.sum()  # Total instances in the matrix
accuracy = true_positives / total_predictions if total_predictions > 0 else 0

# Plot confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix (Boulders and Craters Only)')
plt.show()

# Display overall accuracy
print(f"Overall Accuracy (Boulders and Craters): {accuracy:.4f} ({true_positives}/{total_predictions})")
```
